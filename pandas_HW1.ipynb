```python
ggplot(huron,aes(x='year',y='level',color='decade')) +\
  geom_point(alpha=0.5)+\
  geom_line(alpha=0.5,linetype='solid',size=1) +\
  scale_y_continuous(breaks = np.arange(575,590,0.5),limits = (575,585))+\
  scale_x_continuous(breaks = np.arange(1875,1975,5),limits = (1875,1975))+\
  labs(x='YEAR',
       y='level\n(feet)',
       title='Annual measurements of the level of Lake Huron\n(1875–1972)',
       caption='(Base on data from Brockwell, P. J. and Davis, R. A. (1991). Time Series and Forecasting Methods. Second edition. Springer, New York. Series A, page 555\nBrockwell, P. J. and Davis, R. A. (1996). Introduction to Time Series and Forecasting. Springer, New York. Sections 5.1 and 7.6.)\nDate:'+ 
       pd.Timestamp('today').strftime('%Y/%m/%d') +
       '\nDrawer:Jade Yang')+\
  theme_bw()+\
  theme(figure_size=(12,12),
       text=element_text(fontproperties=rig),
       title=element_text(size=15),
        axis_text=element_text(fontproperties=noto,color='gray'),
        legend_text=element_text(fontproperties=noto))
```
